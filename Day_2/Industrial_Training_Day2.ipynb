# Day 2 matplotlib nd pandas intro

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("IMG20260502202342.jpeg")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
img = Image.open(r"IMG20260502202342.jpeg")   #relative path

img = np.array(img)

img[:,:,0] = np.clip(img[:,:,0] + 98, 100 , 55)

plt.imshow(img)
plt.axis("off")
plt.show()
```
