```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img=Image.open("/content/hello.webp")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
img=Image.open("/content/hello.webp")
img=np.array(img)
img[:,:,0]=np.clip(img[:,:,0]+ 0,0,0)
plt.imshow(img)
plt.axis("off")
plt.show()
```
